## 다변수

### x(설명 변수) / y(목표 변수) : 범주형 / 연속형

```python
df_TOCC.columns
```

```python
df_TOCC['amount'].describe()
```

```python
sns.barplot(data=df_TOCC, x='channel', y='amount') # 기본 평균
plt.show()
```

```python
sns.barplot(data=df_TOCC, x='channel', y='amount', estimator=np.sum) # 기본 총합
plt.show()
```

```python
df_TOCC_bottom_list = pd.pivot_table(data=df_TOCC, index='channel', values='amount', aggfunc=np.sum).sort_values(by='amount', ascending=False).iloc[-7:].index.to_list()
```

```python
df_TOCC_bottom_list
```

```python
sns.barplot(data=df_TOCC, x='channel', y='amount', estimator=np.sum, order=df_TOCC_bottom_list, hue='type_of_contract') # 하위 7개 순위 표현
```
