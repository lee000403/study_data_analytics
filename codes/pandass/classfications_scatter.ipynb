```python
class1_x_value = [0, 1, 2, 3, 4]
class1_y_value = [1, 3, 5, 7, 9]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=class1_x_value, y=class1_y_value)
```

```python
class2_x_value = [1.5, 2.5, 3.5, 4.5, 5.5]
class2_y_value = [3.5, 5.5, 7.5, 9.5, 11.5]
```

```python
# 두 클래스 변수를 한 plt
plt.scatter(x=class2_x_value, y=class2_y_value, label='Class One')
plt.scatter(x=class1_x_value, y=class1_y_value, label='Class Two')
plt.title('Classfications Plot')
plt.xlabel('x Axis') # plt.xlabel('x 축')
plt.ylabel('y Axis') # plt.ylabel('y 축')
plt.legend()
plt.show()
```
