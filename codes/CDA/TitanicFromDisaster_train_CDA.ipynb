```python
df_TOCC_sex_pivot = pd.pivot_table(data=df_TOCC_sex, index='Sex', columns='Survived', aggfunc=len)
df_TOCC_sex_pivot
```

```python
stats.chi2_contingency(observed=df_TOCC_sex_pivot, correction=True)
```
